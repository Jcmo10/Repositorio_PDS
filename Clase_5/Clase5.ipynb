# Convolucion

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir señal 1 y señal 2
signal1 = np.array([0, 1, 2, 1, 0, -1, -2, -1, 0, 1])
signal2 = np.array([1, 1, 0, -1, 2])

# Calcular la convolución entre las dos señales
convolved_signal = np.convolve(signal1, signal2, 'valid')

# Crear una figura y subplots
fig = plt.figure()
ax1 = fig.add_subplot(3, 1, 1)
ax2 = fig.add_subplot(3, 1, 2)
ax3 = fig.add_subplot(3, 1, 3)

ax1.stem(signal1, use_line_collection=True)
ax1.set_title('Señal 1')

ax2.stem(signal2, use_line_collection=True)
ax2.set_title('Señal 2')

ax3.stem(convolved_signal, use_line_collection=True)
ax3.set_title('Señal 1 convolucionada con Señal 2')

ax1.set_xlabel('Muestras')
ax1.set_ylabel('Amplitud')
ax1.set_xlim([0, len(signal1)-1])
ax1.set_ylim([-3, 3])

ax2.set_xlabel('Muestras')
ax2.set_ylabel('Amplitud')
ax2.set_xlim([0, len(signal2)-1])
ax2.set_ylim([-3, 3])

ax3.set_xlabel('Muestras')
ax3.set_ylabel('Amplitud')
ax3.set_xlim([0, len(convolved_signal)-1])
ax3.set_ylim([-3, 3])

plt.tight_layout()
plt.show()
```
