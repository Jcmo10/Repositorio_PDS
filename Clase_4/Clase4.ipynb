# Se√±al escalada en Amplitud 

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la se√±al original
t = np.linspace(0,np.pi , 50)
x = np.sin(2*np.pi*t)

# Escalar la se√±al
k = 2
m = 0.5
y = k * np.sin(2*np.pi*t)

plt.plot(t, x, label='Original')
plt.plot(t, y, label='Escalada')
plt.legend()
plt.tight_layout()
plt.show()

# Graficar ambas se√±ales
plt.subplot(1,2,1)
plt.stem(t, x)
plt.title('Original')
plt.subplot(1,2,2)
plt.stem(t, y)
plt.title('Escalada')
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')

plt.tight_layout()
plt.show()
```

# Desplazamiento de una se√±al

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la se√±al original
t = np.linspace(0, 2, 1000)
x = np.sin(2*np.pi*t)

# Desplazar la se√±al
T = 0.25
y = np.sin(2*np.pi*(t - T))

# Graficar ambas se√±ales
plt.plot(t, x, label='Original')
plt.plot(t, y, label='Desplazada')
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la se√±al original
t = np.linspace(-1, 2, 50)
x = np.sin(2*np.pi*t)

# Desplazar la se√±al
T = 0.25
y = np.sin(2*np.pi*(t - T))


y2 = np.sin(2*np.pi*(t + T))

# Graficar ambas se√±ales
plt.subplot(3,1,1)
plt.stem(t, x, label='Original')
plt.legend()
plt.ylabel('Amplitud')

plt.subplot(3,1,2)
plt.stem(t, y, label='Desplazada -')
plt.legend()
plt.ylabel('Amplitud')

plt.subplot(3,1,3)
plt.stem(t, y2, label='Desplazada +')
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')

plt.tight_layout()
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir se√±al digital
signal = np.array([0, 1, 2, 1, 0, -1, -2, -1, 0, 1])

# Definir tama√±o y calcular se√±al filtrada para este caso se tomo de a 2 datos pero puede ser un filtro para mas dat
window_size = 2
weights = np.repeat(1.0, window_size) / window_size
filtered_signal = np.convolve(signal, weights, 'valid')

# Crear figura y subplots
fig = plt.figure()
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

# Graficar se√±al original en el primer subplot
ax1.stem(signal, use_line_collection=True)
ax1.set_title('Se√±al original')
ax1.grid()

# Graficar se√±al filtrada en el segundo subplot
ax2.stem(filtered_signal, use_line_collection=True)
ax2.set_title('Se√±al filtrada')
ax2.grid()

# Ajustar etiquetas de los ejes y l√≠mites
ax1.set_xlabel('Muestras')
ax1.set_ylabel('Amplitud')
ax1.set_xlim([0, len(signal)-1])
ax1.set_ylim([-2.2, 2.2])

ax2.set_xlabel('Muestras')
ax2.set_ylabel('Amplitud')
ax2.set_xlim([0, len(filtered_signal)-1])
ax2.set_ylim([-2, 2])

# Mostrar figura
plt.tight_layout()
plt.show()

```
