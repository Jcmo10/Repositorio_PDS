## ruido blanco a una señal analógica.

```python
import numpy as np
import matplotlib.pyplot as plt

# Frecuencia de la señal
frecuencia = 1

# Amplitud de la señal
amplitud = 1

# Duración de la señal
duracion = 1

# Número de puntos de muestreo
muestras = 1000

# Tiempo de muestreo
t = np.linspace(0, duracion, muestras)

# Generar la señal sinusoidal
señal = amplitud * np.sin(2 * np.pi * frecuencia * t)

# Graficar la señal
plt.plot(t, señal)

plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt # libreria usada para graficas
# para que la grafica se incruste en el notebook
%matplotlib inline
f=1.0 # Frecuencia de la senal
fs=50 # Frecuencia de muestreo
t=np.arange(0, 2.0, 1.0/fs) # Vector de tiempo
x = np.sin(2*np.pi*f*t)

plt.plot(t,x)
plt.stem(t,x)
plt.xlabel('Time',fontsize=18)
plt.ylabel('Amplitude',fontsize=18)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Frecuencia de la señal
frecuencia = 1

# Amplitud de la señal
amplitud = 1

# Duración de la señal
duracion = 1

# Número de puntos de muestreo
muestras = 1000

# Tiempo de muestreo
t = np.linspace(0, duracion, muestras)

# Generar la señal sinusoidal
señal = amplitud * np.sin(2 * np.pi * frecuencia * t)

# Generar ruido blanco
ruido = 0.1 * np.random.randn(muestras)

# Agregar el ruido a la señal
señal_con_ruido = señal + ruido

# Graficar la señal con ruido
plt.plot(t,señal, color='red', linewidth=3.0)
plt.plot(t, señal_con_ruido)
# Establecer los límites de los ejes x e y
plt.xlim(0, 0.5)
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.show()
```

## Nyquist

```python
import numpy as np
import matplotlib.pyplot as plt

# Frecuencia de la señal
f = 1.0

# Frecuencias de muestreo
fs = [2.0, 4.0, 6.0, 8.0]

# Definir los vectores de tiempo y la señal para cada frecuencia de muestreo
t = []
x = []
for i in range(len(fs)):
    t.append(np.arange(0, 2.0, 1.0/fs[i]))
    x.append(np.sin(2*np.pi*f*t[i]))
    
    # Graficar la señal
    plt.plot(t[i], x[i], label='fs = '+str(fs[i])+' Hz')

# Añadir etiquetas a los ejes y leyenda
plt.xlabel('Time',fontsize=18)
plt.ylabel('Amplitude',fontsize=18)
plt.legend()

# Mostrar la imagen
plt.show()
```
