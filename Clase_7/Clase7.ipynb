# polos y ceros de la funci√≥n de transferencia y la circunferencia unitaria

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Definir funci√≥n de transferencia H(z)
num = [1, -0.5, 0.25]
den = [1, 0.2, 0.3]

sys1 = signal.lti(num, den)
sys = signal.TransferFunction(num, den)

# Obtener polos y ceros
zeros = sys.zeros
poles = sys.poles

# Crear figura y subplot
fig, ax = plt.subplots()

# Graficar c√≠rculo unitario
unit_circle = plt.Circle((0, 0), 1, fill=False, linestyle='--')
ax.add_artist(unit_circle)

# Graficar polos y ceros
ax.scatter(np.real(zeros), np.imag(zeros), marker='o', s=100, label='Ceros')
ax.scatter(np.real(poles), np.imag(poles), marker='x', s=100, label='Polos')

ax.set_xlim([-1.5, 1.5])
ax.set_ylim([-1.5, 1.5])
ax.set_aspect('equal', adjustable='box')
ax.legend()

ax.grid()
plt.show()
```
