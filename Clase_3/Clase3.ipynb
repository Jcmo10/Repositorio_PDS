```python
import numpy as np
import matplotlib.pyplot as plt
import random

secuencia = []
#Se genera una secuencia random de int entre -10 y 10
for i in range(10):
    secuencia.append(random.randint(-10, 10))


n = np.arange(len(secuencia))                 


plt.figure()
plt.stem(n, secuencia)
plt.xlabel('Numero de muestras')
plt.ylabel('Amplitud')
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Definir los valores de "a" y "b"
a = 1
b = 0.5

# Generar algunos puntos en el eje x
x = np.linspace(0, 5, 50)

# Calcular los valores correspondientes de la funci칩n en el eje y
y = a * np.exp(-b*x)

# Crear el gr치fico
plt.stem(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Funci칩n exponencial decreciente')
plt.grid(True)
plt.show()
```

# Se침ales que podrian servir


```python
import matplotlib.pyplot as plt
import numpy as np

# Generar una onda triangular
t = np.linspace(0, 2*np.pi, 50)
x = np.abs((t - np.pi) / np.pi - 1)

# Graficar la se침al
plt.stem(t, x)
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.title('Onda triangular')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generar un pulso cuadrado
t = np.linspace(-1, 2, 40)
x = np.zeros_like(t)
x[(t >= 0.0) & (t <= 1.0)] = 1

# Graficar la se침al
plt.stem(t, x)
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.title('Pulso cuadrado')
plt.show()
```
