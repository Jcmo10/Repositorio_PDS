```python
import numpy as np
import matplotlib.pyplot as plt

# Definir señal 1 y señal 2
signal1 = np.array([1, 2, 3, 0, 3, 1, 1, 2, 1])
signal2 = np.array([1, 0, 1, 0, 1, 0, 1, 0, 1])

# Calcular la correlación entre las dos señales
correlated_signal = np.correlate(signal1, signal2, 'full')


fig = plt.figure()
ax1 = fig.add_subplot(3, 1, 1)
ax2 = fig.add_subplot(3, 1, 2)
ax3 = fig.add_subplot(3, 1, 3)

# Graficar señal 1 en el primer subplot
ax1.stem(signal1, use_line_collection=True)
ax1.set_title('Señal 1')

# Graficar señal 2 en el segundo subplot
ax2.stem(signal2, use_line_collection=True)
ax2.set_title('Señal 2')

# Graficar señal correlacionada en el tercer subplot
ax3.stem(correlated_signal, use_line_collection=True)
ax3.set_title('Correlación entre Señal 1 y Señal 2')


ax1.set_xlabel('Muestras')
ax1.set_ylabel('Amplitud')
ax1.grid()

ax2.set_xlabel('Muestras')
ax2.set_ylabel('Amplitud')
ax2.grid()

ax3.set_xlabel('Retardo')
ax3.set_ylabel('Amplitud')
ax3.grid()


plt.tight_layout()
plt.show()
```
